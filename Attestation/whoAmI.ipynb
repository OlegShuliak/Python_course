```python
import random
import pandas as pd

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
print(data.head(20))
```

```python
pd.get_dummies(data['whoAmI'])
```

```python
from sklearn.preprocessing import OneHotEncoder

onehot_encoder = OneHotEncoder(sparse_output=False) 
onehot_encoded = onehot_encoder.fit_transform(data) 
one_hot = pd.DataFrame(onehot_encoded)
print(one_hot.head(20)) 


```
